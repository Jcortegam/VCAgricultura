```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
im=cv.imread('cat.jpg')
green=im[:,:,1]
plt.imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
plt.show()
```

```python
im.shape
```

```python
plt.imshow(cv.cvtColor(green, cv.COLOR_BGR2RGB))
plt.show()
```

```python
green=green/255
```

```python

vertical=np.array([[-1,-2,-1],
                  [0,0,0],
                  [1,2,1]])
vertical
```

```python
horizontal=np.array([[1,0,-1],
                     [2,0,-2],
                     [1,0,-1]])
horizontal
```

```python
imVer=cv.filter2D(green,-1,vertical)
```

```python
plt.imshow(imVer, cmap="gray")
plt.show()
```

```python
imHor=cv.filter2D(green,-1,horizontal)
```

```python
plt.imshow(imHor,cmap="gray")
plt.show()
```

```python
import math
```

```python
sobelo= np.sqrt(imHor**2 + imVer**2) 
```

```python
plt.imshow(sobelo,cmap="gray")
plt.show()
```
